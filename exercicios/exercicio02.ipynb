### 7. Supondo que o valor da tarifa no dataset está em libras, e que 1£ = R$ 6.43, crie uma nova coluna chamada tarifa_real que contenha a tarifa de embarque em reais.

```python
def tariff(tarifa):
    ganancia = tarifa * 6.43
    return ganancia



df["tarifa_real"] = df["tarifa"].apply(tariff)

df.head(10)
```

### 11. Agrupe pela classe e pela coluna sobreviveu e conte a quantidade de pessoas que sobreviveu de acordo com cada classe. Quem sobreviveu mais? A classe mais alta ou a mais baixa?

```python
grupi = df[["classe","sobreviveu"]] # objeto de classe e sobreviveu
grupi_1 = grupi[((grupi["sobreviveu"]) == "sim") & (grupi["classe"])] # grupo formado por pessoas que sobreviveram
grupi_1
```

```python
grupi_1.groupby("classe").count() # na classe 1, 136 pessoas sobreviveram e na classe 2, 119 pessoas sobreviveram
```

### 12. Agrupe pelo sexo e pela coluna sobreviveu e conte a quantidade de pessoas que sobreviveram de acordo com o sexo. Quem sobreviveu mais? Homens ou mulheres?

```python
grupii = df[["sexo","sobreviveu"]] 
grupi_3 = grupii[((grupii["sexo"]) == "feminino") & (grupii["sobreviveu"])]
grupi_4 = grupii[((grupii["sexo"]) == "masculino") & (grupii["sobreviveu"])]  
grupi_3
```

```python
# separação dentro do sexo feminino onde nos dará o resultado de quantas mulheres sobreviveram e quantas não
grupi_3.groupby("sobreviveu").count()
```

```python
# separação dentro do sexo masculino onde nos dará o resultado de quantos homens sobreviveram e quantas não.
grupi_4.groupby("sobreviveu").count()
```
